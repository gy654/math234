```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
from scipy.stats import norm
```

```python
alpha = 0.05
n = 100
r = norm.rvs(size=n)
```

```python
bound = math.sqrt((1 / (2 * n)) * math.log(2 / alpha)) 
# compute the bound for confidence interval

F_n = lambda x : sum(r < x) / n 
# empirical CDF at x with respect to 100 data points
L_n = lambda x : max(F_n(x) - bound, 0)
# estimated lower bound of CI at x
U_n = lambda x : min(F_n(x) + bound, 1)
# estimated upper bound of CI at x

r_sorted = sorted(r)
# 100 data points from smallest to largest

df = pd.DataFrame({
    'x': r_sorted, 
    'F_n': np.array(list(map(F_n, r_sorted))), 
    # the function F_n maps 100 sorted data points to 100 corresponding values in expirical CDF
    'U_n': np.array(list(map(U_n, r_sorted))), 
    # the function U_n maps 100 sorted data points to 100 corresponding values in upper bound
    'L_n': np.array(list(map(L_n, r_sorted))), 
    # the function L_n maps 100 sorted data points to 100 corresponding values in lower bound
    'CDF': np.array(list(map(norm.cdf, r_sorted)))
})

plt.plot( 'x', 'F_n', data=df, color='blue')
plt.plot( 'x', 'L_n', data=df, color='red')
plt.plot( 'x', 'U_n', data=df, color='green')
plt.plot( 'x', 'CDF', data=df, color='black')
plt.legend();
```

1000 iterations with normal distribution

```python
inbound = []
for k in range(1000):
    n = 100
    alpha = 0.05
    r = norm.rvs(size=n)
    bound = math.sqrt((1 / (2 * n)) * math.log(2 / alpha))

    F_n = lambda x : sum(r < x) / n
    L_n = lambda x : max(F_n(x) - bound, 0)
    U_n = lambda x : min(F_n(x) + bound, 1)

    r_sorted = r
    
    df = pd.DataFrame({
        'x': r_sorted, 
        'F_n': np.array(list(map(F_n, r_sorted))), 
        'U_n': np.array(list(map(U_n, r_sorted))), 
        'L_n': np.array(list(map(L_n, r_sorted))), 
        'CDF': np.array(list(map(norm.cdf, r_sorted)))
    })
    all_in_bounds =  ((df['L_n'] <= df['CDF'])&(df['CDF'] <=df['U_n'] )).all() 
    # examine each of 100 data points on true CDF, if in bound, mark 1, otherwise mark 0.

    inbound.append(all_in_bounds)

    # inbound list now contains 1000 indicator lists with size 100.
    
print("the frequency that the confidence band contains the true distribution function is:", np.array(inbound).mean())
```

An estimator for $\theta$ would be $\hat{\theta} = \hat{p_{1}}-\hat{p_{2}}$
where $\hat{p_{1}}$ = number of people recovered under the standard treatment/ total number of people in the first group. 
$\hat{p_{2}}$ = number of people recovered under the new treatment/ total number of people in the second group. 

$ Var(\hat{\theta})=E[\hat{\theta}^{2}]-E[\hat{\theta}]^{2} = E[\hat{p_{1}}^{2}+\hat{p_{2}}^{2}-2\hat{p_{1}}\hat{p_{2}}]-(p_{1}-p_{2})^{2} = \frac{p_{1}-p_{1}^{2}+ p_{2}-p_{2}^{2}}{n}$

$ se(\hat{\theta})=\sqrt{\frac{\hat{p_{1}}-\hat{p_{1}}^{2}+ \hat{p_{2}}-\hat{p_{2}}^{2}}{n}} $

```python

p1hat = 90/100
p2hat = 85/100
n = 100
theta_hat = p1hat - p2hat
se_theta_hat = math.sqrt((p1hat * (1 - p1hat) + p2hat * (1 - p2hat)) / n)
print("An estimator theta_hat would be ", theta_hat)
print("The standard error of this estimator would be ", se_theta_hat)
```

```python
from scipy.stats import norm

z_80 = norm.ppf(0.9)
z_95 = norm.ppf(0.975)
upperBound_80 = theta_hat + z_80 * se_theta_hat
lowerBound_80 = theta_hat - z_80 * se_theta_hat
print("The 80 percent confidence interval is ({}, {}) ".format(lowerBound_80, upperBound_80))
upperBound_95 = theta_hat + z_95 * se_theta_hat
lowerBound_95 = theta_hat - z_95 * se_theta_hat
print("The 95 percent confidence interval is ({}, {}) ".format(lowerBound_95, upperBound_95))

```

```python
import numpy as np
from scipy.stats import norm
```

 generate data set X from standard normal distribution, let Y be generated from $ Y_{i} = e^{X_{i}} $

```python

def generate(n):
    X = norm.rvs(size = n)
    Y = np.exp(X)
    return Y
```

calculate $T(F)= \int \frac{(x-\mu)^{3}dF(x)}{\sigma^{3}}$

```python
def T(Y):
    n = len(Y)
    mu = sum(Y)/n
    var = sum((Y - mu)**2) / n
    skewness = sum((Y - mu)**3)/(n*var**(3/2))
    return skewness
```

calculate bootstrap estimator of $10^{4}$ trials and store the result in a $(10^{4},1)$ array $\Theta^{*}$

```python
def bootstrap(Y, Ne = 10**4):
    n = len(Y)
    t_bootstrap = np.empty(Ne)
    # create an array to store bootstrap estimators
    for i in range(Ne):
        y_star = np.random.choice(Y, n, replace=True)
        # Yi_star are drawn from Y with replacement
        t_bootstrap[i] = T(y_star)
        # calculate T(F)for each bootstrap sample and store the estimator in t_bootstrap array
    return t_bootstrap
        
```

calculate 3 types of confidence interval given $\Theta^{*}$, $\hat{\theta}$, and $\alpha$

```python
def bootstrapInterval (t_bootstrap, theta_hat, alpha = 0.05):
    se_boot = t_bootstrap.std()
    # se_boot is also the sqare root of v_boot = E[T(Y*)|Y1, Y2,...Yn]
    z = norm.ppf(1 - alpha/2)
    # get z value for confidence level alpha
    
    normal_CI = (theta_hat - z * se_boot, theta_hat + z * se_boot)
    
    q_half_alpha = np.quantile(t_bootstrap, alpha/2)
    q_c_half_alpha = np.quantile(t_bootstrap, 1 - alpha/2)
    
    pivotal_CI = (2* theta_hat - q_c_half_alpha, 2*theta_hat - q_half_alpha)
    
    percentile_CI = (q_half_alpha, q_c_half_alpha)
    
    return normal_CI, pivotal_CI, percentile_CI
    
```

Notes:

normal confidence interval: $(\hat{\theta} - z_{1-\frac{\alpha}{2}}\cdot \hat{se}_{boot},\hat{\theta} + z_{1-\frac{\alpha}{2}} \cdot \hat{se}_{boot})$

pivotal confidence interval: $(2\hat{\theta}-\theta^{*}_{\frac{1-\alpha}{2}}, 2\hat{\theta}-\theta^{*}_{\frac{\alpha}{2}})$, where $\theta^{*}(\alpha)$ is $\alpha $ quatile of $\theta$

percentile confidence interval: $(\theta^{*}_{\frac{\alpha}{2}},  \theta^{*}_{\frac{1-\alpha}{2}})$

```python
n = 50
Y = generate(n)
theta_hat = T(Y) # estimator of Y from which bootstrap samples are drawn
t_bootstrap = bootstrap(Y, Ne = 10**4) # an array of bootstrap estimators
normal_CI, pivotal_CI, percentile_CI= bootstrapInterval(t_bootstrap, theta_hat, alpha = 0.05)
```

```python
print("The normal confidence interval for bootstrap is: ",normal_CI)
print("The pivotal confidence interval for bootstrap is: ", pivotal_CI)
print("The percentile confidence interval for bootstrap is: ", percentile_CI)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm
```

```python
def theta_cdf(x):
    return norm.cdf(np.log(x), loc = 5, scale = 1/100)
```

```python
n = 100
X = np.random.normal(5,1, size = n) 
# X is drawn from a normal distribution with mu=5, variance =1

X_bar = np.mean(X)
theta_hat = np.exp(X_bar)

```

```python
Ne = 10**4
t_boot = np.empty(Ne)
n = len(X)
for i in range (len(t_boot)):
    x_star = np.random.choice(X, n, replace = True)
    t_boot[i]= np.exp(np.mean(x_star))
    
```

```python
se_boot = t_boot.std()
print("The se for the bootstrap is: ", se_boot)
alpha = 0.05
z = norm.ppf(1-alpha/2)
lowerBound = theta_hat - z*se_boot
upperBound = theta_hat + z*se_boot
CI = (lowerBound, upperBound)
print("The 95 percent confidence interval i: ", CI)
```

Additional problems 1.

we implement the bootstrap for a correlation estimation problem.

Let $W,Z \sim N(0,1)$ be independent. Define $X = W + \gamma Z$ and $Y = Z + \gamma W$.

Then
$\mathbb{E}[X] = \mathbb{E}[Y] = 0, \quad \mathbb{E}[X^2] = \mathbb{E}[Y^2] = 1 + \gamma^2, \quad \mathbb{E}[XY] = 2\gamma, $ 

so
$\rho_{X,Y} = \frac{\mathbb{E}[XY]}{\sqrt{ \mathbb{E}[X^2]\, \mathbb{E}[Y^2] }}  = \frac{2 \gamma }{1+\gamma^2}.
$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfinv
from scipy.stats import norm
```

```python
gamma = 0.1        # model parameter
alpha = .05   # confidence interval parameter
N = 100   # number of bootstrap process per experiment
Ne=1000 # number of experiments
# exact correlation, the equation is from the notebook
rho_exact = (2*gamma)/(1+gamma**2)
```

```python
# create multiple sets for future plot
n_list=[]
rho_hat_list=[] #  analytical rhohat averaged over Ne experiments related to sample size n
t_boot_list=[]  # bootstrap estimator averaged over Ne experiments related to sample size n
for n in range(5,505,10):
    n_list.append(n) # list of sample size n 
    rho_hat_exp=[] # list containing analytical rhohat for each experiment
    t_boot_exp=[] # list containing bootstrap estimator for each expiriment
    for i in range(Ne):  # for different independent experiment
        # we generate the data for each experiment
        W = norm.rvs(0,1,n)
        Z = norm.rvs(0,1,n)
        X = W + gamma*Z
        Y = Z + gamma*W
        # compute the analytical rhohat for each experiment
        rho_hat_exp.append(np.mean(X*Y) / np.sqrt( np.mean(X*X) * np.mean(Y*Y)))
        # bootstrap process
        t_boot=np.empty(N) # N bootstrap samples
        for j in range (N):
            I=np.random.choice(n,n)
            xx=X[I]
            yy=Y[I]
            t_boot[j] = np.mean(xx*yy) / np.sqrt( np.mean(xx*xx) * np.mean(yy*yy) )
        t_boot_exp.append(t_boot.mean()) # record the mean value of bootstrap estimator for each experiment(t_boot_exp)
    t_boot_list.append(np.mean(t_boot_exp)) # record the t_boot experiment to a list relating to sample size n 
    rho_hat=np.mean(rho_hat_exp)
    rho_hat_list.append(rho_hat)
# compute bias for different n values
```

```python
bias=[]
for i in range(50):
    bias.append(t_boot_list[i]-rho_hat_list[i])
# compute the bias corrected theta hat
bias_corrected=[]
for i in range(50):
    bias_corrected.append(rho_hat_list[i]-bias[i])
# bias for corrected theta hat
bias_corrected_bias=[]
for i in range(50):
    bias_corrected_bias.append(bias_corrected[i]-rho_exact)
# add RMSE for the corrected theta hat
bias_corrected_RMSE=[]
for i in range(50):
    bias_corrected_RMSE.append(np.sqrt(np.mean((bias_corrected[i] - rho_exact)**2)))

# bias for plug in estimator
plugin_bias=[]
for i in range(50):
    plugin_bias.append(rho_hat_list[i]-rho_exact)
# the RMSE for plug in estimator
plugin_RMSE=[]
for i in range(50):
    plugin_RMSE.append(np.sqrt(np.mean((rho_hat_list[i] - rho_exact)**2)))

    
```

```python
# plot (plug in estimator)
plt.plot(n_list,plugin_bias,color='red',label='bias of plug-in theta hat')
plt.plot(n_list,plugin_RMSE,color='blue',label='RMSE of plug-in theta hat')
plt.legend()
plt.show()
```

```python
# plot(corrected_ for bias)
plt.plot(n_list,bias_corrected_bias,color='green',label='bias of bias-corrected theta hat')
plt.plot(n_list,bias_corrected_RMSE,color='purple',label='RMSE of bias-corrected theta hat')
plt.legend()
plt.show()
```
